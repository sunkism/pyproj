```python
import sys
import pandas as pd
import numpy as np

# INPUT FILE
in_file = 'NYSE_H_20191217.E2.csv'

# CSV파일을 읽는다
df = pd.read_csv(in_file, sep=',',dtype='unicode', parse_dates=True,skiprows=1)

# 필요없는 컬럼 지우기
df.drop([df.columns[0],df.columns[3],df.columns[4],df.columns[5],df.columns[6]],axis=1,inplace=True)

# date 중 뒤에 '.1' 붙어있는 것 제거
for i in df.columns:
    if ".1" == i[-2:]:
        df.rename(columns={i : i[:-2]}, inplace=True)

# 날짜컬럼 type 변경 (string -> datetime)
df.columns = df.columns[:2].tolist() + pd.to_datetime(df.columns[2:]).strftime('%Y-%m-%d').tolist()

# date컬럼들을 행으로바꾸기
temp = df.melt(id_vars=["INSTRUMENT(MNEMONIC)", "DATATYPES"], 
        var_name="Date", 
        value_name="Value")

# DATATYPES를 컬럼으로 바꾸기
result = temp.set_index(['INSTRUMENT(MNEMONIC)','Date','DATATYPES']).unstack(level=-1)

out_file = 'convert_'+in_file
# CSV 저장
result.to_csv(out_file, mode='w')

print ('** FILE CREATED : ' + out_file)
```
