```python
import sys
import pandas as pd
import numpy as np

# full number 보이게 하기
pd.options.display.float_format = "{:.f}".format

in_file = 'historydata.xlsx'

print ("program start!")
df = pd.read_excel(in_file,dtype='unicode')
print ("end!")
```

```python
cols = df.columns.drop(['DATE','CODE','NAME'])
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
df1 = df.drop('NAME',axis=1)
df1.set_index(['DATE','CODE'], inplace=True)
df1 = df1.unstack(level=1)
df1.columns.set_names('INFO',level = 0,inplace=True)
```

```python
#색인방법

# 멀티레벨 컬럼 색인. 
A = df1['PRC','000070'][:'20150107'] 
B = df1[:][:'20150120']

# loc로 할땐 row , column 순서!
C = df1.loc[:'20150120']['PRC','000070']

# 멀티레벨 컬럼에서 선택하기
idx = pd.IndexSlice
# level 0: 모든값, level 1: '0000070' 만
D = df1.loc[:,idx[:,'000070']]
# level 0: 모든값, level 1: '0000070', '024720' 
E = df1.loc[:,idx[:,['000070','024720']]]
# level 0: PRC, SHRS,   level 1: '000070','024720'
F = df1.loc[:,idx[['PRC','SHRS'], ['000070','024720'] ]]
# 20150120일까지 데이터만, level 0: PRC, SHRS,   level 1: '000070','024720'level 0: PRC, SHRS,   level 1: '000070','024720'
G = df1.loc[:'20150120',idx[['PRC','SHRS'], ['000070','024720'] ]]
```

```python
# 시가총액 계산
# 계산은 쉽다. but...만든결과를 df1에 추가하는게 좀 까다롭다. multilevel columns를 이용한게 원인인듯하다.
# Multilevel column이라 그런지..일반적인 방식으론 새로운 열을 df1에 바로 추가할 수 없다. ㅜㅜ
# 우선 df1과 동일한 column정보를 가지도록 MKTCAP을 구하고,
# concat으로 합친다...
# 좀더 효율적인 방식은 나중에 알아보자

# 우선 MKTCAP 계산
MKTCAP = df1['PRC'].mul(df1['SHRS'])
# 위의 결과는 Multilevel columns가 아니다. 그냥 종목코드만 컬럼
# 종목코드는 level 1로 두고, 'MKTCAP'을 level 0로 세팅!
MKTCAP.columns = pd.MultiIndex.from_product([['MKTCAP'],MKTCAP.columns])

# df1, MKT의 컬럼명도 동일하게!
MKTCAP.columns.set_names(df1.columns.names, inplace=True)

# concat을 이용하여 병합!
df1 = pd.concat([df1,MKTCAP],axis=1)
```
